```python
import pandas as pd

print(pd.__version__)
data = pd.read_csv('happyscore_income.csv')
```

```python
income = data['avg_income']
happy = data.happyScore
ineq = data['income_inequality']

happy.max()
income.min()
#print(happy, income)
```

```python
import matplotlib.pyplot as plt

plt.scatter(income, happy)
plt.xlabel('Average Income')
plt.ylabel('Happy Score')
plt.show()
```

```python
plt.scatter(income, ineq)
plt.xlabel('Average Income')
plt.ylabel('Income Inequality')
plt.show()
```

```python
plt.scatter(ineq, happy)
plt.xlabel('Income Inequality')
plt.ylabel('Happy Score')
plt.show()
```

```python
plt.scatter(income, happy, s=ineq*10, alpha=0.25) # s mean size. small size circle mean low income inequality
plt.xlabel('Average Income')
plt.ylabel('Happy Score')
plt.show()
```

```python
data.sort_values('avg_income', inplace=True) # Sorting pandas data frame

riches = data[ data['avg_income'] > 15000 ] # filtering pandas data frame

data # print table like structure
riches.iloc[-1] # print last row
data.iloc[0:5, -2] # print 2nd last column of first 5 rows
data.iloc[0:5, -2:] # print last 2 columns of first 5 rows
```

```python
import numpy as np

rich_mean = np.mean(riches['avg_income'])
rich_std = np.std(riches['avg_income'])
print(rich_mean, rich_std)

all_mean = np.mean(data['avg_income'])
all_std = np.std(data['avg_income'])
print(all_mean, all_std)
```

```python
riches_income = riches['avg_income']
riches_happy = riches['happyScore']
riches_country = riches['country']

plt.scatter(riches_income, riches_happy)
plt.xlabel("Average Income", fontsize=14, color='red')
plt.ylabel("Happy Score", fontsize=14, color='red')

plt.text(riches.iloc[0]['avg_income'], # x position
         riches.iloc[0]['happyScore'], # y position
         riches.iloc[0]['country'])    # label to print on graph

plt.show()
```

```python
plt.scatter(riches_income, riches_happy)
plt.xlabel("Income", fontsize=14, color='red')
plt.ylabel("Happy Score", fontsize=14, color='red')

for x,y,z in zip(riches_income, riches_happy, riches_country):
    label=z # riches.iloc[0]['country']
    plt.annotate(label,
                (x,y),
                #xytext=(0,1), # distance from text to points (x,y)
                ha='center') # horizontal alignment can be left, right or center

plt.show()
```

```python
plt.scatter(riches_income, riches_happy)
plt.xlabel("Income", fontsize=14, color='red')
plt.ylabel("Happy Score", fontsize=14, color='red')

for k,row in riches.iterrows(): # k is the index or key of each row
    plt.text(row['avg_income'],
             row['happyScore'],
             row['country'])

# This is list comprehension. It can also work for labelling
#[plt.text(x=row['avg_income'], y=row['happyScore'], s=row['country']) for k,row in data.iterrows()]

plt.show()
```
