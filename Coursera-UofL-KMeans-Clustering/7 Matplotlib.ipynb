```python
import numpy as np

xys = [[10, 20], [45, 92], [120, 40], [67, 100], [82, 60]]
xys = np.array(xys)
print(xys[0,0])
print(xys[0,1]) # -1 can be used instead of 1 as -1 gives last columnb
print(xys[1,0])
print(xys[1,1])
print(xys)
```

```python
plt.subplot(3,3,9)
plt.scatter(xys[:,0], xys[:,1])
plt.show()
```

```python
#mean = np.mean(xys) # flattens all data and calculates mean. Wrong way
mean = np.mean(xys, 0) # preserves data structure. 0 -> column wise mean, 1 -> row wise mean
std_dev = np.std(xys, 0)
print(xys, mean, std_dev)
xys.mean(0)
```

```python
plt.scatter(xys[:,0], xys[:,1])
plt.scatter(mean[0], mean[1])
plt.show()
```

```python
# to add graphical patches
import matplotlib.patches as patches

mean = np.mean(xys, 0)
std_dev = np.std(xys, 0) # same as mean. 0 -> column wise std dev, 1 -> row wise std dev

ellipse = patches.Ellipse([mean[0], mean[1]], std_dev[0]*2, std_dev[1]*2, alpha=0.25) # alpha is for transparancy

fig,graph = plt.subplots()

graph.scatter(xys[:,0], xys[:,1])
graph.scatter(mean[0], mean[1])
graph.add_patch(ellipse)
```

```python
dist = np.linalg.norm(xys[0]-mean) # to calculate Euclidean distance of 1st point from mean
print(dist)
```

```python
# Euclidean distance of each point from mean
dists = [np.linalg.norm(xy-mean) for xy in xys]
print(dists)
```

```python
print(xys)

x_min = np.min(xys[:,0])
x_max = np.max(xys[:,0])

x_normed = (xys[:,0] - x_min) / (x_max - x_min)
print(x_min, x_max, x_normed)
```

```python
# Normalisation of multidimensional data
xys_min = np.min(xys, 0) # same as mean and std
xys_max = np.max(xys, 0)

normed = (xys - xys_min) / (xys_max - xys_min)
print(normed)
```

```python
# plot normed data
mean = np.mean(normed, 0)
std_dev = np.std(normed, 0) # same as mean. 0 -> column wise std dev, 1 -> row wise std dev

ellipse = patches.Ellipse([mean[0], mean[1]], std_dev[0]*2, std_dev[1]*2, alpha=0.25) # alpha is for transparancy

fig,graph = plt.subplots()

graph.scatter(normed[:,0], normed[:,1])
graph.scatter(mean[0], mean[1])
graph.add_patch(ellipse)
```
