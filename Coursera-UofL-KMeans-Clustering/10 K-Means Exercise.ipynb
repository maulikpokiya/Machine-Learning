```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

data = pd.read_csv('happyscore_income.csv')
```

```python
region = data['region']
gdp = data['GDP']
happy = data['happyScore']
```

```python
plt.xlabel('GDP')
plt.ylabel('Happy Score')

plt.scatter(gdp, happy)
plt.show()
```

```python
data.sort_values('GDP', inplace=True)
gdp = data['GDP']
happy = data['happyScore']

plt.xlabel('GDP')
plt.ylabel('Happy Score')

plt.plot(gdp, happy)
plt.show()
```

```python
riches = data[ data['avg_income'] > 15000 ] 
poors = data[ data['GDP'] > 1 ] 

gdp = poors['GDP']
happy = poors['happyScore']

happy_gdp = np.column_stack((gdp, happy))
km_res = KMeans(n_clusters=2).fit(happy_gdp)

clusters = km_res.cluster_centers_
labels = km_res.labels_
print(clusters, labels)
```

```python
plt.xlabel('GDP')
plt.ylabel('Happy Score')

plt.scatter(gdp, happy) # plot graph points
plt.scatter(clusters[:,0], clusters[:,1], s=1000, alpha=0.5) # plot cluster centers

plt.text(poors.iloc[3]['GDP'], # x position
         poors.iloc[3]['happyScore'], # y position
         poors.iloc[3]['country'])    # label to print on graph

#for k,row in poors.iterrows(): # k is the index or key of each row
#    plt.text(row['GDP'],
#             row['happyScore'],
#             row['country'])

plt.show()
```
